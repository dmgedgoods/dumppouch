```python
import folium
import pandas as pd
from folium.plugins import HeatMap

# Load your data
data = pd.read_csv("C:/Users/shawn/code/python/SFDP2018.csv")

# Create a base map
map_center = [data['latitude'].mean(), data['longitude'].mean()]
my_map = folium.Map(location=map_center, zoom_start=10)

# Prepare heat map data
heat_data = [[row['latitude'], row['longitude'], row['row_id']] for index, row in data.iterrows()]

# Add heat map layer
HeatMap(heat_data).add_to(my_map)

# Display the map
my_map
```

```python
import folium
import pandas as pd
from folium.plugins import HeatMap, MarkerCluster

# Load the dataset
file_path = 'C:/Users/shawn/code/python/SFDP2018.csv'
df = pd.read_csv(file_path)

# Define the initial map center based on average coordinates
map_center = [37.7749, -122.4194]
my_map = folium.Map(location=map_center, zoom_start=12)

# Prepare heatmap data
heat_data = [[row['latitude'], row['longitude']] for index, row in df.iterrows()]

# Add HeatMap layer
HeatMap(heat_data).add_to(my_map)

# Initialize marker cluster
marker_cluster = MarkerCluster().add_to(my_map)

# Add markers to the cluster
for index, row in df.iterrows():
    folium.Marker(location=[row['latitude'], row['longitude']],
                  popup=f"Category: {row['incident_subcategory']}\nDescription: {row['incident_description']}\nResolution: {row['resolution']}").add_to(marker_cluster)

# Display the map
my_map
```

```python
import folium
import pandas as pd
from folium.plugins import HeatMap, MarkerCluster

# Load the dataset
file_path = 'C:/Users/shawn/code/python/SFDP2018.csv'  # Ensure the correct file path
df = pd.read_csv(file_path)

# Take a random sample of the data
sample_df = df.sample(n=10000, random_state=42)

# Define the initial map center based on average coordinates
map_center = [sample_df['Latitude'].mean(), sample_df['Longitude'].mean()]
my_map = folium.Map(location=map_center, zoom_start=12)

# Prepare heatmap data
heat_data = sample_df[['Latitude', 'Longitude']].dropna().values.tolist()

# Add HeatMap layer
HeatMap(heat_data).add_to(my_map)

# Save the map to an HTML file
my_map
```
